# Python Classes & Inheritance - Day 7

## 1. What is a Class?
A class is basically a "blueprint" or a "template" for creating things.
It's like a recipe for a cake — the recipe itself isn't a cake, but you use it to bake as many real cakes as you want!

## 2. Why use a Class?
In programming, we use classes to group together data (variables) and actions (functions) into one object.
Think of it like a player profile in a video game — one profile holds your name, your level, your health, and your abilities all in one place.

## 3. How to write a Class
Use the `class` keyword and then the name of the class (usually capitalized).
Inside, we use a special function called `__init__` to set up our template.
Think of `__init__` like the assembly line at a car factory — every time a new car rolls off the line, it gets stamped with a color, a model, and a serial number.

`self` refers to the specific object we are creating right now.

```python
class Dog:
    def __init__(self, dogName, dogBreed):
        self.dogName = dogName
        self.dogBreed = dogBreed

    def barkAtStranger(self):
        print(self.dogName + " says Woof! Who are you?")

myDog = Dog("Buddy", "Golden Retriever")
myDog.barkAtStranger()
```

## 4. What is Inheritance?
Inheritance lets a new class "borrow" everything from an existing class.
It works just like a child inheriting traits from a parent — you don't have to teach a child to walk from scratch if they already picked it up naturally. They just add their own skills on top.

If we want a specific type of dog, like a PoliceDog, we don't want to rewrite all the `dogName` and `bark` code. We just inherit from the Dog class.

```python
class PoliceDog(Dog):  # parent class in parentheses
    def sniffForClues(self):
        print(self.dogName + " is looking for clues!")

k9 = PoliceDog("Max", "German Shepherd")
k9.barkAtStranger()   # Inherited from Dog
k9.sniffForClues()    # New skill
```

## 5. Mutability
Objects created from a class are mutable — that means you can change their attributes after they've been created.
Think of it like a whiteboard. Once something is written on it, you can erase it and write something new.

This is different from a String, which is immutable. Once a String is created, Python won't let you change individual parts of it — like text carved into stone.

```python
myDog.dogName = "Max"  # changed Buddy's name to Max
print(myDog.dogName)
```

```python
dogOne = Dog("Buddy", "Golden Retriever")
dogTwo = Dog("Rex", "Labrador")

dogOne.dogName = "Charlie"
print(dogOne.dogName)  # Charlie
print(dogTwo.dogName)  # Rex (completely unaffected)
```

## Summary
- Class: The Blueprint (Dog)
- Object: The real thing made from the blueprint (myDog)
- Inheritance: Getting features from a Parent class (PoliceDog inheriting from Dog)
- Mutability: Objects can be changed after creation. Each object owns its own data.
